```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

classifier(["Ahmed Abohadeed Is A Great AI Engineer" , "Ahmed Abohadeed Need Money"])
```

## Zero-shot-classification pipline

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "Ahmed Abohadeed is Learning Transformers",
    candidate_labels=["education", "politics", "business"],
)
```

## The Text-generation-pipeline

```python
generator = pipeline("text-generation" , model="distilgpt2")
generator("Ahmed Abohadeed is a great AI Engineer" ,

          max_length=30,
          num_return_sequences=2,
          )
```

```python
unmasker = pipeline("fill-mask")
unmasker("Ahmed Abohadeed is A Great AI <mask>" , top_k=4)
```

## The Ner pipline

```python
from transformers import pipeline

ner = pipeline("ner", aggregation_strategy="simple")
ner("Ahmed Abohadeed is A Great AI Engineer Work as CEO of HADEEd Vision")
```
